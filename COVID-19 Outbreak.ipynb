### Predict using Moving Average

```python
confirm_data_group.rolling(window=1, axis=1).mean()
```

```python
confirm_data_group.ewm(span=1).mean()
```

```python
def cal_moving_avg(input_df, window=2):
    """Calculates Moving avg using the window size."""
    _df = input_df.rolling(window, axis=1).median()['3/10/20']
    
    return _df

def cal_ewma(input_df, comm=0.3):
    """Calculates the exp wighted moving average using the window size."""
    _df = input_df.ewm(com=comm).mean()['3/10/20']
    
    return _df
```

```python
all_pred_df = pd.DataFrame()
all_pred_df_ew = pd.DataFrame()

for df in [confirm_data_group, recover_data_group, death_data_group]:
    _df = cal_moving_avg(df, window=2)
    ew_df = cal_ewma(df, comm=0.4)
    all_pred_df = pd.concat([all_pred_df, _df], axis=1)
    all_pred_df_ew = pd.concat([all_pred_df_ew, ew_df], axis=1)
    
all_pred_df.columns =['Confirmed', 'Recovered', 'Death']    
all_pred_df_ew.columns =['Confirmed', 'Recovered', 'Death']
```

### We can use a function to convert all the 3 csv files to our required format

```python
all_csv = [cnf_data, death_data, recover_data]
```

```python
def convert_by_country_csv(dataframe, files):
    
    #Transposing
    dataframe = dataframe.transpose()
    
    #Taking Country Names as Cols
    cols = dataframe.iloc[0,:]
    
    #Making cols as the columns of the df
    dataframe.columns = list(cols)
    
    #Dropping 1st row as they are country names
    dataframe.drop(dataframe.index[0], inplace=True)
    
    #Export to Separate csv file
    dataframe.to_csv(files[0]+'.csv')#Naming csv file as per Cases
    del files[0]
    #print(files)
```

```python
files = ['Confirmed_cases_country', 'Deaths_country', 'Recovered_country']
```

```python
for csvfile in all_csv:
    convert_by_country_csv(csvfile,files)
```
